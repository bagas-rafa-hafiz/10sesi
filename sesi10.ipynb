```python
import numpy as np
import pandas as pd

n_data = 10000


days = np.random.choice(['Senin','Selasa','Rabu','Kamis','Jumat','Sabtu','Minggu'], size=n_data)
date = np.random.randint(1,31, size=n_data)
Jam = np.random.randint(00.10,12.00, size=n_data)
Deposit = np.random.choice(['50.000','100.000','500.000'], size=n_data)
nama = np.random.choice(['btc', 'bnb', 'mtc'], size=n_data)
transaksi = np.random.randint(200.000,600.000,size=n_data)
jumlah = np.random.randint(1,1000, size=n_data)
Buy = np.random.choice(['100.000','1.555.345.355','2.000.000'],size=n_data)
Sell = np.random.randint(100.000, 500.000, size=n_data)
Penarikan = np.random.choice(['100.000','500.000','2.000.000'],size=n_data)

df = pd.DataFrame({
    'days' : days,
    'date' : date,
    'Jam' : Jam,
    'Deposit' : Deposit,
    'Coin' : nama,
    'Transaksi' : transaksi,
    'Jumlah' : jumlah,
    'Buy' : Buy,
    'Sell' : Sell,
    'Penarikan' :Penarikan
})

df.to_csv('dataLatih.csv',index=False)

df.head ()# untuk mengetahui 5 baris pertama dari isi dataframe
```
